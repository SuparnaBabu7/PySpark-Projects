# 29) Write a query to find which products are most preferred by customers based on the total sales. Display customer name, favorite product (top 3 products per each customer), and total sales on that product? 

```python
#29) Write a query to find which products are most preferred by customers based on the 
# total sales. Display customer name, favorite product (top 3 products per each customer), and total sales on that product? 
from pyspark.sql import Window
df_t1.join(df_t2,df_t1["OrderID"] == df_t2["OrderID"],"inner").groupBy("CustomerName","ProductName")\
     .agg(sum("Sales").alias("TotalSales")).sort(desc("TotalSales"))\
     .withColumn("Rank",row_number().over(Window.partitionBy("CustomerName").orderBy(desc("TotalSales"))))\
     .filter(col("Rank") <=3).select("CustomerName","ProductName","TotalSales").show()
```

# 30) Write a query to get 7th rank customer name based on total sales? display customer name, sales amount & rank

```python
# 30) Write a query to get 7th rank customer name based on total sales? display customer name, sales amount & rank
df_t1.join(df_t2,df_t1["OrderID"] == df_t2["OrderID"],"inner").groupBy("CustomerName")\
     .agg(sum("Sales").alias("TotalSales")).withColumn("Rank",dense_rank().over(Window.orderBy(desc("TotalSales"))))\
     .filter(col("Rank") == 7).select("CustomerName","TotalSales","Rank").show()
```
