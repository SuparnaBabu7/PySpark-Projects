```python
from pyspark.sql.functions import * 
```

```python
df = spark.read.format('delta').load("abfss://344326ff-0c2a-460d-ab0c-21ad5eebd0ab@onelake.dfs.fabric.microsoft.com/01c4e46e-d603-438d-973b-3f9bb4b0d13d/Tables/ele_store",inferSchema = True)
```

```python
#1) Query the list of columns?
df.columns
```

```python
#2) Query the top 5 rows? 
display(df.head(5))
```

```python
#3) Query the bottom 5 rows?
display(df.tail(5))
```

```python
#4) How many total number of orders placed? 
orders = df.select("OrderID").count()
print("Total orders placed :",orders)
```

```python
#5) What are the unique channels present and how many are there? 
df.select("Channel").distinct().show()
channel = df.select("Channel").distinct().count()
print("Total number of Channels :",channel)
```

```python
#6) How many cities are there? 
cities = df.select("City").distinct().count()
print("Total number of cities :",cities)
```

```python
#7) Query the unique Countries and count of the countries?  
countries = df.select("Country").distinct()
display(countries)
countries_count = df.select("Country").distinct().count()
print("The number of countries :", countries_count)
```

```python
#8) Query the unique Regions and count of the regions?
regions = df.select("Region").distinct()
display(regions)
regions_count = df.select("Region").distinct().count()
print("The number of regions :",regions_count)
```

```python
#9) Query the total cities count by Country and order it from highest count to lowest count?
df.groupBy("Country").agg(countDistinct("City").alias("Count of Cities")).sort("Count of Cities", ascending = False).show()
```

```python
#10) How many products are there? Display count of products 
products_count = df.select("ProductName").distinct().count()
print("The number of products :",products_count)
products = df.select("ProductName").distinct()
display(products)
```

```python
#11) Query the total sales and total profit? 
sales = df.agg(sum("Sales").alias("Total Sales"))
profit = df.agg(sum("Profit").alias("Total Profit"))
display(sales)
display(profit)
```

```python
#12) Query the total sales by region and country? 
sales = df.groupBy("Region","Country").agg(sum("Sales").alias("Total Sales"))
display(sales)
```

```python
#13) Query the total order qty by promotion name? 
orderqty = df.groupBy("PromotionName").agg(sum("OrderQty").alias("Total Order Qty"))
display(orderqty)
```

```python
#14) Query the top 7 countries with sales from highest to lowest order? 
top7countries_sales = df.groupBy("Country").agg(sum("Sales").alias("Total Sales")).sort("Total Sales", ascending = False).limit(7)
display(top7countries_sales)
```

```python
#15) How much the total sales happened in Asia region? 
asiasales = df.filter(col("Region") == "Asia").agg(sum("Sales").alias("Asia_Sales"))
display(asiasales)
```

```python
#16) How much total sales happened in Asia region and India? 
asiaindidasales = df.filter((col("Region") == "Asia") & (col("Country") == "India")).agg(sum("Sales").alias("Asia_India_Sales"))
display(asiaindidasales)
```

```python
#17)  How much profit got in Europe region and Denmark country? 
europe_denmark_profit = df.filter((col("Region") == "Europe") & (col("Country") == "Denmark"))\
                          .agg(sum("Profit").alias("Total Profit"))
display(europe_denmark_profit)
```

```python
#18)  Query the total order qty by Manufacturer and Product Name? 
manf_prod_qty = df.groupBy("Manufacturer","ProductName").agg(sum("OrderQty").alias("Total Order Qty"))\
                  .sort("Total Order Qty", ascending = False)
display(manf_prod_qty)
```

```python
#19) Query the top 7 products with orderqty and sales and show it descending order based on total orderqty? 
top7products = df.groupBy("ProductName").agg(sum("OrderQty").alias("Total Order Qty"),sum("Sales").alias("Total Sales"))\
                 .sort("Total Order Qty", ascending = False)
display(top7products)
```

```python
#20)  Query the total order quantity greater than 10000 by city?
city_10k = df.groupBy("City").agg(sum("OrderQty").alias("Total Order Qty")).filter(col("Total Order Qty") > 10000)
display(city_10k)
```

```python
# 21) Query the total sales less than 1000 by product, display it in ascending order?
product_less_1k = df.groupBy("ProductName").agg(sum("Sales").alias("Total Sales")).filter(col("Total Sales") < 1000)\
                    .sort("Total Sales", ascending = True)
display(product_less_1k)
```

```python
#22) Query the total profit greater than 200000 by product, display it in descending order? 
profit_greater_2lakh = df.groupBy("ProductName").agg(sum("Profit").alias("Total Profit")).filter(col("Total Profit") > 200000)\
                         .sort("Total Profit", ascending = False)
display(profit_greater_2lakh)
```

```python
#23)  How much total sales happened in China and Beijing?
chi_bjng_sales = df.filter((col("Country") == "China") & (col("City") == "Beijing")).agg(sum("Sales").alias("Total Sales"))
display(chi_bjng_sales)
```

```python
#24)  How much total sales happened in Asian Holiday Promotion?
as_holi_promo = df.filter(col("PromotionName") == "Asian Holiday Promotion").agg(sum("Sales").alias("Total Sales"))
display(as_holi_promo)
```

```python
#25)  How much total profit value by Contoso, Ltd Manufacturer? 
conto_profit = df.filter(col("Manufacturer") == "Contoso, Ltd").agg(sum("Profit").alias("Total Profit"))
display(conto_profit)
```

```python
#26)  Query the total sales and total order qty by product category and order it by Orderqty from highest to lowest? 
cate_sale_qty = df.groupBy("ProductCategory").agg(sum("Sales").alias("Total Sales"),sum("OrderQty").alias("Total Order Qty"))\
                  .sort("Total Order Qty", ascending = False)
display(cate_sale_qty)
```

```python
#27)  Query the total sales value in between 1057 and 26700 by Region? 
regionsale = df.select("Region","Sales").filter((col("Sales") > 1057) & (col("Sales") < 26700) )
display(regionsale)
```

```python
#28)  How many orders placed in Pittsfield city?
pits_orders = df.filter(col("City") == "Pittsfield").agg(sum("OrderQty").alias("Total Order Qty"))
display(pits_orders)
```

```python
#29)  Query the total sales and total profit by region, country and city?
rcc_sales_profit = df.groupBy("Region","Country","City").agg(sum("Sales").alias("Total Sales"))
display(rcc_sales_profit)
```

```python
#30)  Query the total unit cost and total price by productcategory, productsubcategory and Manufacturer?
cscm_up = df.groupBy("ProductCategory","ProductSubCategory","Manufacturer").agg(sum("UnitCost").alias("Total Unit Cost")\
            ,sum("Price").alias("Total Price"))
display(cscm_up)
```

```python
#converting date data type from string to date data type
df = df.withColumn("OrderDate",to_date(df["OrderDate"],"dd-MM-yyyy"))
```
